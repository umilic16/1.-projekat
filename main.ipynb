## Analiza dataset-a

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv', sep=';')
df = df.rename(columns={
    'Daytime/evening attendance\t':'Daytime/evening attendance',
    'Curricular units 1st sem (credited)': '1st_sem_credited',
    'Curricular units 1st sem (enrolled)': '1st_sem_enrolled',
    'Curricular units 1st sem (evaluations)': '1st_sem_evaluations',
    'Curricular units 1st sem (approved)': '1st_sem_approved',
    'Curricular units 1st sem (grade)': '1st_sem_grade',
    'Curricular units 1st sem (without evaluations)': '1st_sem_wo_evaluations',
    'Curricular units 2nd sem (credited)': '2nd_sem_credited',
    'Curricular units 2nd sem (enrolled)': '2nd_sem_enrolled',
    'Curricular units 2nd sem (evaluations)': '2nd_sem_evaluations',
    'Curricular units 2nd sem (approved)': '2nd_sem_approved',
    'Curricular units 2nd sem (grade)': '2nd_sem_grade',
    'Curricular units 2nd sem (without evaluations)': '2nd_sem_wo_evaluations'
})
df
```

## Priprema i predobrada podataka

### Podela dataseta

```python
X = df.drop('Target', axis=1)
y = df['Target']
```

### SMOTE balansiranje

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)

X_bal, y_bal = smote.fit_resample(X, y)
```

### Konvertovanje odredjenih kolona u binarne

```python
# X['1st_sem_wo_evaluations'] = (X['1st_sem_wo_evaluations'] > 0).astype(int)
# X['2nd_sem_wo_evaluations'] = (X['2nd_sem_wo_evaluations'] > 0).astype(int)
```

### Konvertovanje kategorickih vrednosti

```python
categorical_cols = ['Marital status', 'Application mode', 'Course', 'Previous qualification', 'Nacionality', 'Mother\'s qualification', 'Father\'s qualification', 'Mother\'s occupation', 
                    'Father\'s occupation']


X_enc = pd.get_dummies(X_bal, columns=categorical_cols, dtype=int)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y_enc = label_encoder.fit_transform(y_bal)
```

```python
X_enc.info()
```

### Train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_enc, y_enc, test_size=0.2)
```

### Transformacija numerickih vrednosti

```python
# numeric_cols = ['Application order', 'Previous qualification (grade)', 'Admission grade', 'Age at enrollment', '1st_sem_credited', '1st_sem_enrolled', '1st_sem_evaluations', '1st_sem_approved',
#                 '1st_sem_grade', '2nd_sem_credited', '2nd_sem_enrolled', '2nd_sem_evaluations', '2nd_sem_approved', '2nd_sem_grade', 'Unemployment rate', 'Inflation rate', 'GDP']
numeric_cols = ['Application order', 'Previous qualification (grade)', 'Admission grade', 'Age at enrollment', '1st_sem_credited', '1st_sem_enrolled', '1st_sem_evaluations', '1st_sem_approved',
                '1st_sem_grade','1st_sem_wo_evaluations', '2nd_sem_credited', '2nd_sem_enrolled', '2nd_sem_evaluations', '2nd_sem_approved', '2nd_sem_grade','2nd_sem_wo_evaluations', 'Unemployment rate', 'Inflation rate', 'GDP']
```

```python
X_train[numeric_cols].hist(figsize=(15,15))
```

```python
y_train_df = pd.DataFrame(y_train, columns=['Target'])
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(X_train[numeric_cols].join(y_train_df).corr(), annot=True, cmap='YlGnBu', fmt='.2f')
```

#### Standardizacija

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
# X_train_std = X_train_no_outliers.copy()
X_train_std = X_train.copy()
X_train_std[numeric_cols] = standard_scaler.fit_transform(X_train_std[numeric_cols])

X_test_std = X_test.copy()
X_test_std[numeric_cols] = standard_scaler.fit_transform(X_test_std[numeric_cols])
```

```python
X_train_std[numeric_cols].hist(figsize=(15,15))
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(X_train_std[numeric_cols].join(y_train_df).corr(), annot=True, cmap='YlGnBu', fmt='.2f')
```

## Kreiranje modela

### Dummy classifier

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy='uniform', random_state=1)

dummy.fit(X_train_std, y_train)
```

```python
accuracy_dummy = dummy.score(X_test_std, y_test)
accuracy_dummy
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression(max_iter=1000, multi_class='multinomial')
logreg.fit(X_train_std, y_train)
y_predict = logreg.predict(X_test_std)

accuracy_log = accuracy_score(y_test, y_predict)
accuracy_log
```

```python
from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test, y_predict)
dataframe = pd.DataFrame(matrix, index=logreg.classes_, columns=logreg.classes_)
sns.heatmap(dataframe, annot=True, cmap="Blues", fmt='d')
plt.title("Confusion Matrix - Logistic Regression"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=16)
knn.fit(X_train_std.values, y_train)
y_predict = knn.predict(X_test_std.values)

accuracy_knn = accuracy_score(y_test, y_predict)
accuracy_knn
```

```python
matrix = confusion_matrix(y_test, y_predict)
dataframe = pd.DataFrame(matrix, index=knn.classes_, columns=knn.classes_)
sns.heatmap(dataframe, annot=True, cmap="Blues", fmt='d')
plt.title("Confusion Matrix - KNN"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

decisiontree = DecisionTreeClassifier(random_state=8)
decisiontree.fit(X_train_std, y_train)
y_predict = decisiontree.predict(X_test_std)
```

```python
accuracy_dt = accuracy_score(y_test, y_predict)
accuracy_dt
```

```python
matrix = confusion_matrix(y_test, y_predict)
dataframe = pd.DataFrame(matrix, index=knn.classes_, columns=knn.classes_)
sns.heatmap(dataframe, annot=True, cmap="Blues", fmt='d')
plt.title("Confusion Matrix - Decision Tree"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

### Random forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=6, n_jobs=-1)
model = rfc.fit(X_train_std, y_train)
y_predict = rfc.predict(X_test_std)

accuracy_rfc = accuracy_score(y_test, y_predict)
accuracy_rfc
```

```python
matrix = confusion_matrix(y_test, y_predict)
dataframe = pd.DataFrame(matrix, index=knn.classes_, columns=knn.classes_)
sns.heatmap(dataframe, annot=True, cmap="Blues", fmt='d')
plt.title("Confusion Matrix - Random Forest Classifier"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

### Support Vector Classification

```python
from sklearn.svm import LinearSVC

svc = LinearSVC(dual="auto")
svc.fit(X_train_std, y_train)
y_predict = svc.predict(X_test_std)

accuracy_svc = accuracy_score(y_test, y_predict)
accuracy_svc
```

```python
matrix = confusion_matrix(y_test, y_predict)
dataframe = pd.DataFrame(matrix, index=knn.classes_, columns=knn.classes_)
sns.heatmap(dataframe, annot=True, cmap="Blues", fmt='d')
plt.title("Confusion Matrix - SVC"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

## Uporedjivanje modela

```python
classifiers = ['Dummy', 'Logistic Regression', 'KNN', 'Decision tree', 'Random Forest', 'SVC']

accuracy = [accuracy_dummy, accuracy_log, accuracy_knn, accuracy_dt, accuracy_rfc, accuracy_svc]

norm = plt.Normalize(min(accuracy), max(accuracy))
colors = plt.cm.Blues(accuracy)

plt.figure(figsize=(10, 6))
plt.bar(classifiers, accuracy, color=colors)
plt.ylabel('Accuracy')
plt.xlabel('Classifiers')
plt.title('Comparison of accuracies')
plt.ylim(0, 1) 
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Kombinovanje feature-a iz 1. i 2. semestra u 1. godinu

```python
columns_to_sum = [
    '1st_sem_credited', '1st_sem_enrolled', '1st_sem_evaluations',
    '1st_sem_approved', '1st_sem_grade', '1st_sem_wo_evaluations',
    '2nd_sem_credited', '2nd_sem_enrolled', '2nd_sem_evaluations',
    '2nd_sem_approved', '2nd_sem_grade', '2nd_sem_wo_evaluations'
]

X_train2['1st_year_credited'] = X_train2['1st_sem_credited'] + X_train2['2nd_sem_credited']
X_train2['1st_year_enrolled'] = X_train2['1st_sem_enrolled'] + X_train2['2nd_sem_enrolled']
X_train2['1st_year_evaluations'] = X_train2['1st_sem_evaluations'] + X_train2['2nd_sem_evaluations']
X_train2['1st_year_approved'] = X_train2['1st_sem_approved'] + X_train2['2nd_sem_approved']
X_train2['1st_year_grade'] = X_train2['1st_sem_grade'] + X_train2['2nd_sem_grade']
X_train2['1st_year_wo_evaluations'] = X_train2['1st_sem_wo_evaluations'] + X_train2['2nd_sem_wo_evaluations']

X_train2 = X_train2.drop(columns=columns_to_sum, axis=1)

X_test2['1st_year_credited'] = X_test2['1st_sem_credited'] + X_test2['2nd_sem_credited']
X_test2['1st_year_enrolled'] = X_test2['1st_sem_enrolled'] + X_test2['2nd_sem_enrolled']
X_test2['1st_year_evaluations'] = X_test2['1st_sem_evaluations'] + X_test2['2nd_sem_evaluations']
X_test2['1st_year_approved'] = X_test2['1st_sem_approved'] + X_test2['2nd_sem_approved']
X_test2['1st_year_grade'] = X_test2['1st_sem_grade'] + X_test2['2nd_sem_grade']
X_test2['1st_year_wo_evaluations'] = X_test2['1st_sem_wo_evaluations'] + X_test2['2nd_sem_wo_evaluations']

X_test2 = X_test2.drop(columns=columns_to_sum, axis=1)
```
