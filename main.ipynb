## Priprema i predobrada podataka

### Feature engineering

```python
df['Marital status'] = df['Marital status'].eq(1).astype(int)
```

```python
def map_application_mode(mode):
    if mode in [1, 17, 18]:
        # 'General Contingent'
        return 0
    elif mode in [7, 10, 15, 42, 43, 44, 51, 53, 57]:
        # 'Special Contingent'
        return 1
    elif mode == 39:
        # 'Over 23 Years Old'
        return 3
    else:
        return 4

df['Application mode'] = df['Application mode'].apply(map_application_mode)
```

```python
def map_course(course_code):
    if course_code in [33, 171]:
        # 'Technology and Engineering'
        return 0
    elif course_code in [9500, 9556, 9085]:
        # 'Healthcare'
        return 1
    elif course_code in [9147, 9991, 9670]:
        # 'Management and Business'
        return 2
    elif course_code in [9070, 171]:
        # 'Design and Multimedia'
        return 3
    elif course_code in [9003, 9130]:
        # 'Agriculture and Environment'
        return 4
    elif course_code in [8014, 9238]:
        # 'Social Services'
        return 5
    elif course_code == 9254:
        # 'Tourism'
        return 6
    elif course_code == 9853:
        # 'Education'
        return 7
    else:
        return 8

df['Course'] = df['Course'].apply(map_course)
```

```python
df['Previous qualification'] = df['Previous qualification'].eq(1).astype(int)
```

```python
df['Nacionality'].value_counts()
```

```python
df.drop(columns=['Nacionality'], inplace=True)
```

```python
def map_qualification(qualification):
    if qualification == 1:
        # 'Secondary Education'
        return 1
    elif qualification in [19, 37, 38]:
        # 'Basic Education'
        return 2
    else:
        return 3

df['Mother\'s qualification'] = df['Mother\'s qualification'].apply(map_qualification)
df['Father\'s qualification'] = df['Father\'s qualification'].apply(map_qualification)
```

```python
def map_occupation(occupation):
    occupation_mapping = {
        # 'Specialists in Intellectual and Scientific Activities'
        2: 0,
        # 'Intermediate Level Technicians and Professions'
        3: 1,
        # 'Administrative staff'
        4: 2,
        # 'Personal Services, Security and Safety Workers and Sellers'
        5: 3,
        # 'Skilled Workers in Industry, Construction and Craftsmen'
        7: 4,
        # 'Unskilled Workers'
        9: 5
    }
    
    return occupation_mapping.get(occupation, 6)

df['Mother\'s occupation'] = df['Mother\'s occupation'].apply(map_occupation)
df['Father\'s occupation'] = df['Father\'s occupation'].apply(map_occupation)
```

```python
df['Educational special needs'].value_counts()
```

```python
df.drop(columns=['Educational special needs'], inplace=True)
```

```python
def map_age(age):
    if 17 <= age <= 21:
        return 0
    elif 22 <= age <= 25:
        return 1
    elif 26 <= age <= 30:
        return 2
    elif 31 <= age <= 35:
        return 3
    elif 36 <= age <= 40:
        return 4
    elif 41 <= age <= 50:
        return 5
    else:
        return 6

df['Age at enrollment'] = df['Age at enrollment'].apply(map_age)
```

```python
df['International'].value_counts()
```

```python
df.drop(columns=['International'], inplace=True)
```

```python
df['1st_sem_credited'] = df['1st_sem_credited'].eq(1).astype(int)
df['1st_sem_wo_evaluations'] = df['1st_sem_wo_evaluations'].eq(1).astype(int)
df['2nd_sem_credited'] = df['2nd_sem_credited'].eq(1).astype(int)
df['2nd_sem_wo_evaluations'] = df['2nd_sem_wo_evaluations'].eq(1).astype(int)
```

### Podela dataseta

```python
X = df.drop('Target', axis=1)
y = df['Target']
```

### SMOTE balansiranje

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)

X_bal, y_bal = smote.fit_resample(X, y)
```

### Konvertovanje kategorickih vrednosti

```python
categorical_cols = ['Application mode', 'Course',  'Mother\'s qualification', 'Father\'s qualification', 'Mother\'s occupation', 'Father\'s occupation']


X_enc = pd.get_dummies(X_bal, columns=categorical_cols, dtype=int)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y_enc = label_encoder.fit_transform(y_bal)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_enc, y_enc, test_size=0.2)
```

## Kreiranje modela

### Dummy classifier

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy='uniform', random_state=1)

dummy.fit(X_train_std, y_train)

accuracy_dummy = dummy.score(X_test_std, y_test)
accuracy_dummy
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

logreg = LogisticRegression(max_iter=1000, multi_class='multinomial')
logreg.fit(X_train_std, y_train)
y_predict = logreg.predict(X_test_std)

accuracy_log = accuracy_score(y_test, y_predict)

report = classification_report(y_test, y_predict)
print(report)
```

```python
importance = logreg.coef_[0]
en_imp = [(i, v) for i, v in enumerate(importance)]
sorted_en_imp = sorted(en_imp, key=lambda x: x[1], reverse=True)

for i, v in sorted_en_imp:
    print(f"{X_test_std.columns[i]}, Value: {v}")
```

```python
from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test, y_predict)
dataframe = pd.DataFrame(matrix, index=logreg.classes_, columns=logreg.classes_)
sns.heatmap(dataframe, annot=True, cmap="Blues", fmt='d')
plt.title("Confusion Matrix - Logistic Regression"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=16)
knn.fit(X_train_std.values, y_train)
y_predict = knn.predict(X_test_std.values)

accuracy_knn = accuracy_score(y_test, y_predict)

report = classification_report(y_test, y_predict)
print(report)
```

```python
matrix = confusion_matrix(y_test, y_predict)
dataframe = pd.DataFrame(matrix, index=knn.classes_, columns=knn.classes_)
sns.heatmap(dataframe, annot=True, cmap="Blues", fmt='d')
plt.title("Confusion Matrix - KNN"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```
